```python
from langchain_experimental.graph_transformers import LLMGraphTransformer
from tqdm import tqdm
import time
from typing import List
from langchain.docstore.document import Document

def process_documents_in_batches(
    documents: List[Document],
    llm_transformer: LLMGraphTransformer,
    batch_size: int = 5,
    retry_attempts: int = 3,
    delay_between_batches: float = 1.0
) -> List:
    """
    Process documents in batches with retry logic and progress tracking.

    Args:
        documents: List of documents to process
        llm_transformer: The LLM transformer instance
        batch_size: Number of documents to process in each batch
        retry_attempts: Number of retry attempts for failed batches
        delay_between_batches: Delay in seconds between processing batches

    Returns:
        List of processed graph documents
    """
    graph_documents = []

    # Calculate total number of batches
    total_batches = (len(documents) + batch_size - 1) // batch_size

    for i in tqdm(range(0, len(documents), batch_size), total=total_batches, desc="Processing batches"):
        batch = documents[i:i + batch_size]

        for attempt in range(retry_attempts):
            try:
                # Process the current batch
                batch_results = llm_transformer.convert_to_graph_documents(batch)
                graph_documents.extend(batch_results)

                # Add delay between batches to avoid rate limiting
                time.sleep(delay_between_batches)
                break

            except Exception as e:
                if attempt == retry_attempts - 1:
                    print(f"Failed to process batch {i//batch_size + 1} after {retry_attempts} attempts")
                    print(f"Error: {str(e)}")
                else:
                    print(f"Attempt {attempt + 1} failed for batch {i//batch_size + 1}. Retrying...")
                    time.sleep(delay_between_batches * 2)  # Longer delay between retries

    return graph_documents
```

```python

# Usage example:
def process_documents_with_monitoring(documents, llm, batch_size=5):
    """
    Process documents with monitoring and error handling.
    """
    try:
        llm_transformer = LLMGraphTransformer(llm=llm)

        # Print processing information
        print(f"Total documents to process: {len(documents)}")
        print(f"Batch size: {batch_size}")
        print(f"Estimated number of API calls: {len(documents) // batch_size + 1}")

        # Process documents
        graph_documents = process_documents_in_batches(
            documents=documents,
            llm_transformer=llm_transformer,
            batch_size=batch_size
        )

        print(f"\nSuccessfully processed {len(graph_documents)} documents")
        return graph_documents

    except Exception as e:
        print(f"An error occurred during processing: {str(e)}")
        return []
```

```python
from typing import List, Dict, Any
from langchain_google_genai import GoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain.prompts import ChatPromptTemplate
from langchain.chains import LLMChain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Neo4jVector
from langchain.schema import Document
from pydantic import BaseModel, Field
import numpy as np
from tqdm import tqdm
from neo4j import GraphDatabase

class EnhancedRAGSystem:
    def __init__(
        self,
        google_api_key: str,
        neo4j_url: str,
        neo4j_username: str,
        neo4j_password: str
    ):
        """
        Initialize the Enhanced RAG System.
        """
        # Initialize LLM
        self.llm = GoogleGenerativeAI(
            model="gemini-pro",
            google_api_key=google_api_key,
            temperature=0.3
        )

        # Initialize embeddings
        self.embeddings = GoogleGenerativeAIEmbeddings(
            model="models/embedding-001",
            google_api_key=google_api_key,
            task_type="retrieval_query"
        )

        # Initialize Neo4j connection
        self.neo4j_driver = GraphDatabase.driver(
            neo4j_url,
            auth=(neo4j_username, neo4j_password)
        )

        # Initialize vector store
        self.vector_store = Neo4jVector.from_existing_graph(
            embedding=self.embeddings,
            url=neo4j_url,
            username=neo4j_username,
            password=neo4j_password,
            node_label="Document",
            text_node_properties=["text"],
            embedding_node_property="embedding"
        )

        # Initialize prompts
        self.init_prompts()

    def __del__(self):
        """Cleanup Neo4j connection"""
        if hasattr(self, 'neo4j_driver'):
            self.neo4j_driver.close()

    def init_prompts(self):
        """Initialize various prompts used in the system"""
        # Entity extraction prompt
        self.entity_prompt = ChatPromptTemplate.from_messages([
            ("system", """You are an expert at extracting legal entities, acts, and terms from text.
                         Focus on identifying specific laws, regulations, and legal terminology."""),
            ("human", "Extract key legal entities from this text: {text}")
        ])

        # Summarization prompt
        self.summary_prompt = ChatPromptTemplate.from_messages([
            ("system", """Summarize the following legal text, focusing on:
                         1. Key legal points
                         2. Main requirements or obligations
                         3. Important definitions
                         4. Relevant procedures
                         Keep the summary clear and structured."""),
            ("human", "Text to summarize: {text}")
        ])

        # Question answering prompt
        self.qa_prompt = ChatPromptTemplate.from_messages([
            ("system", """You are a legal expert assistant. Use the following context to answer the question.
                         Provide specific references to relevant sections when possible.

                         Context: {context}"""),
            ("human", "Question: {question}")
        ])

    async def extract_entities(self, text: str) -> List[str]:
        """Extract legal entities from text"""
        chain = LLMChain(llm=self.llm, prompt=self.entity_prompt)
        response = await chain.arun(text=text)

        # Parse the response into a list of entities
        entities = [entity.strip() for entity in response.split('\n') if entity.strip()]
        return entities

    async def generate_summary(self, text: str) -> str:
        """Generate a structured summary of the legal text"""
        chain = LLMChain(llm=self.llm, prompt=self.summary_prompt)
        summary = await chain.arun(text=text)
        return summary

    def semantic_search(self, query: str, k: int = 3) -> List[Document]:
        """Perform semantic search using Neo4j vector store"""
        results = self.vector_store.similarity_search(
            query,
            k=k,
            search_type="hybrid"  # Combines semantic and keyword search
        )
        return results

    def _run_cypher_query(self, cypher_query: str, params: Dict = None) -> List[Dict]:
        """Helper method to run Cypher queries"""
        if params is None:
            params = {}

        with self.neo4j_driver.session() as session:
            result = session.run(cypher_query, params)
            return [dict(record) for record in result]

    def structured_retrieval(self, query: str) -> List[Dict[str, Any]]:
        """
        Perform structured retrieval using Neo4j graph relationships
        """
        # Query to find related entities and their relationships
        cypher_query = """
        MATCH (d:Document)-[r]->(e:Entity)
        WHERE d.text CONTAINS $search_query
        RETURN d.text as source, type(r) as relationship, e.name as entity
        LIMIT 5
        """

        return self._run_cypher_query(cypher_query, {"search_query": query})

    async def answer_question(self, question: str) -> Dict[str, Any]:
        """
        Main method to answer questions using RAG
        """
        # 1. Retrieve relevant documents
        semantic_results = self.semantic_search(question)
        structured_results = self.structured_retrieval(question)

        # 2. Combine and process context
        context = "\n".join([
            f"Document {i+1}: {doc.page_content}"
            for i, doc in enumerate(semantic_results)
        ])

        # Add structured information
        context += "\nStructured Information:\n" + "\n".join([
            f"{result['entity']} ({result['relationship']})"
            for result in structured_results
        ])

        # 3. Generate answer
        chain = LLMChain(llm=self.llm, prompt=self.qa_prompt)
        answer = await chain.arun(context=context, question=question)

        # 4. Return comprehensive response
        return {
            "answer": answer,
            "sources": [doc.metadata for doc in semantic_results],
            "structured_info": structured_results
        }

    async def process_document(self, document: str) -> Dict[str, Any]:
        """
        Process a new document for inclusion in the RAG system
        """
        # 1. Extract entities
        entities = await self.extract_entities(document)

        # 2. Generate summary
        summary = await self.generate_summary(document)

        # 3. Store document and entities in Neo4j
        store_query = """
        CREATE (d:Document {text: $doc_text, summary: $doc_summary})
        WITH d
        UNWIND $entity_list as entity
        MERGE (e:Entity {name: entity})
        CREATE (d)-[:MENTIONS]->(e)
        """

        self._run_cypher_query(
            store_query,
            {
                "doc_text": document,
                "doc_summary": summary,
                "entity_list": entities
            }
        )

        return {
            "entities": entities,
            "summary": summary
        }
```

```python
from typing import List, Dict, Any
import asyncio
from langchain_google_genai import GoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain.prompts import ChatPromptTemplate
from langchain.chains import LLMChain
from langchain.vectorstores import Neo4jVector
from neo4j import GraphDatabase

class LegalSearchEngine:
    def __init__(
        self,
        google_api_key: str,
        neo4j_url: str,
        neo4j_username: str,
        neo4j_password: str
    ):
        """Initialize the Legal Search Engine"""
        # Initialize LLM with legal-specific temperature
        self.llm = GoogleGenerativeAI(
            model="gemini-pro",
            google_api_key=google_api_key,
            temperature=0.1  # Lower temperature for more precise legal responses
        )

        # Initialize embeddings
        self.embeddings = GoogleGenerativeAIEmbeddings(
            model="models/embedding-001",
            google_api_key=google_api_key,
            task_type="retrieval_query"
        )

        # Initialize Neo4j connection
        self.neo4j_driver = GraphDatabase.driver(
            neo4j_url,
            auth=(neo4j_username, neo4j_password)
        )

        # Initialize vector store with existing graph
        self.vector_store = Neo4jVector.from_existing_graph(
            embedding=self.embeddings,
            url=neo4j_url,
            username=neo4j_username,
            password=neo4j_password,
            node_label="Document",
            text_node_properties=["text"],
            embedding_node_property="embedding"
        )

        # Initialize legal-specific prompts
        self.init_legal_prompts()

    def __del__(self):
        """Cleanup Neo4j connection"""
        if hasattr(self, 'neo4j_driver'):
            self.neo4j_driver.close()

    def init_legal_prompts(self):
        """Initialize legal-specific prompts"""
        self.qa_prompt = ChatPromptTemplate.from_messages([
            ("system", """You are a legal expert assistant specializing in Indian law.
                         Analyze the following legal context and provide a detailed, structured answer to the question.
                         Include specific sections, rules, and precedents where applicable.
                         Format your response with clear headings and bullet points for better readability.

                         Context: {context}"""),
            ("human", "Question: {question}")
        ])

    def legal_semantic_search(self, query: str, k: int = 5) -> List[Dict]:
        """Perform semantic search optimized for legal queries"""
        # Hybrid search combining vector and keyword search
        results = self.vector_store.similarity_search(
            query,
            k=k,
            search_type="hybrid"
        )

        return results

    def get_related_legal_entities(self, query: str) -> List[Dict]:
        """Retrieve related legal entities and their relationships"""
        cypher_query = """
        MATCH (d:Document)-[r]-(e)
        WHERE d.text CONTAINS $query
        WITH d, r, e
        OPTIONAL MATCH (e)-[r2]-(related)
        RETURN DISTINCT
            d.text as source_text,
            type(r) as relationship_type,
            e.text as entity,
            collect(DISTINCT {
                type: type(r2),
                entity: related.text
            }) as related_entities
        LIMIT 10
        """

        with self.neo4j_driver.session() as session:
            result = session.run(cypher_query, {"query": query})
            return [dict(record) for record in result]

    async def answer_legal_query(self, question: str) -> Dict[str, Any]:
        """
        Main method to answer legal questions using the existing knowledge base
        """
        # 1. Retrieve relevant legal documents
        semantic_results = self.legal_semantic_search(question)
        related_entities = self.get_related_legal_entities(question)

        # 2. Combine context with structured information
        context = "\n\nLegal Documents:\n" + "\n".join([
            f"Document {i+1}: {doc.page_content}\n"
            for i, doc in enumerate(semantic_results)
        ])

        # Add related legal entities and their relationships
        if related_entities:
            context += "\n\nRelated Legal Concepts and Relationships:\n"
            for entity in related_entities:
                context += f"\n {entity.get('entity', '')}"
                if entity.get('related_entities'):
                    for related in entity['related_entities']:
                        if related.get('entity'):
                            context += f"\n  - {related['type']}: {related['entity']}"

        # 3. Generate comprehensive legal answer
        chain = LLMChain(llm=self.llm, prompt=self.qa_prompt)
        answer = await chain.arun(context=context, question=question)

        # 4. Return structured response
        return {
            "answer": answer,
            "sources": [
                {
                    "content": doc.page_content,
                    "metadata": doc.metadata
                } for doc in semantic_results
            ],
            "related_concepts": related_entities
        }

# Example usage
async def main():
    try:
        legal_engine = LegalSearchEngine(
            google_api_key="AIzaSyB75kXU5OSZBp4oycOnjk-uavpJgg537yI",
            neo4j_url="neo4j+s://ffc2cc0f.databases.neo4j.io",
            neo4j_username="neo4j",
            neo4j_password="iH1Qe61EwRwhWtoVncW4XiADuUaABOvKtOagu1NY1m4"
        )

        result = await legal_engine.answer_legal_query(
            "What are the reporting obligations for banks under the Money Laundering Act?"
        )

        print("\nAnswer:", result["answer"])
        print("\nSources:", len(result["sources"]), "documents found")
        print("\nRelated Legal Concepts:", len(result["related_concepts"]), "concepts found")

    except Exception as e:
        print(f"Error: {str(e)}")
    finally:
        if 'legal_engine' in locals():
            del legal_engine


async def run_main():
    await main()

# This will work in Colab
await run_main()
```
